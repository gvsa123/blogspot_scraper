```python
# Should be in script.py???
import nltk
from nltk.corpus import stopwords
```

```python
# Remove stopwords
def remove_stopwords(sent):
    """Remove stopwords"""
    for word in sent:
        if word in stopwords.words('english'):
            sent.remove(word)
    return sent
```

```python
def str_converter(i):
    """String conversion for remove_stopwords"""
    sent_lst = str(i).split(" ")
    sent_lst = remove_stopwords(sent_lst)
    return sent_lst
```

```python
df2 = df.applymap(lambda x: str_converter(str(x).lower()))
```
